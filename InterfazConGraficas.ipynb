```python
import tkinter as tk

from matplotlib.backends.backend_tkagg import (
    FigureCanvasTkAgg, NavigationToolbar2Tk)
#Implementar the las Matplotlib key bindings por defecto
from matplotlib.backend_bases import key_press_handler
from matplotlib.figure import Figure

import matplotlib.pyplot as plt

import numpy as np

from sklearn.datasets.samples_generator import make_blobs

#El papel de ésta variable es «ubicar» en que paso del algoritmo se encuentra el usuario
##Más adelante en las funcines botonSiguiente y botonAnterior
contadorPantallas = 0

#Aquí van los textos que usaremos para cada una de las pantallas
textoCero = "1) Digamos que tenemos un conjunto de datos inicial\ncomo el que se muestra a continuación"
textoUno = "2) Una máquina de soporte vectorial dividirá los datos, pero\n hay infinitas líneas que hacen eso"
textoDos = "3) Entonces una MSV escogerá la que creé el mayor ancho del «camino»"


#Se crea la ventana principal que de aquí en adelante se le llamará raiz
raiz = tk.Tk()

#Se le agrega un título a la raíz
raiz.wm_title("Máquina de soporte vectorial")

#Se crea una figura con acho quince y altura cuatro
figura = Figure(figsize=(8, 4), dpi=100)



#Se parte en intervalo en partes muy pequeñas, esta lista de aquí en adelante se llamará equis
##Esto se usa para que más adelante la gráfica se vea continua
equis, yes = make_blobs(n_samples=50, centers=2,
                  random_state=0, cluster_std=0.60)


subgrafica = figura.add_subplot(111)

#Se le agreda una subgráfica a la figura
##En el eje de las abscisas tendra a equis
##En el eje de las ordenadas tendra las imagen de la función aplicada a equis
subgrafica.scatter(equis[:, 0], equis[:, 1], c=yes, s=50, cmap='autumn');


#Se usa FigureCanvasTkAgg que importamos de matplotlib.backend_bases
##El canvas se agrega a la figura y esta se coloca en raiz
##canvas es una tk.DrawingArea, es decir, es un lugar para graficar
canvas = FigureCanvasTkAgg(figura, master=raiz)  #
#Se «imprime» el canvas
canvas.draw()
#Se agrega como un widget
canvas.get_tk_widget().pack(side=tk.TOP, fill=tk.BOTH, expand=1)



#Se le egrega una barra de herramientas ligada el canvas que estará en la raiz
##De aquí en adelante se le llamara barraHerramientas
##Para esto se usa NavigationToolbar2Tk que importamos de matplotlib.backends.backend_tkagg
barraHerramientas = NavigationToolbar2Tk(canvas, raiz)
barraHerramientas.update()
#Se agrega como un widget la barra de herramientas
canvas.get_tk_widget().pack(side=tk.TOP, fill=tk.BOTH, expand=1)

#Se va a usar una etiqueta en la ventana y con un texto que irá explicando el algoritmo
##El texto sera de color rojo, tipo de letra «Comic Sans MS» y tamaño 18
explicacion = tk.Label(raiz, text=textoCero, fg="white", font=("Comic Sans MS", 18))
##También lo ubicamos
explicacion.pack()
            
#Sinceramente, no se para que sirve está función, si alguien puede investigar y usarla me dice
def on_key_press(evento):
    print("you pressed {}".format(evento.key))
    key_press_handler(evento, canvas, barraHerramientas)

#canvas.mpl_connect("key_press_event", on_key_press)


#Se define la función ligada con el botón para cerrar la ventana
def cerrar():
    raiz.quit()     # Para el mainloop
    raiz.destroy()  # Esto es necesario para prevenir el Fatal Python Error: PyEval_RestoreThread: NULL tstate

    
#Se crea el botón para cerrar la ventana
#Se ubica en al raíz, se el pone como texto «Cerrar» y usa la función cerrar
botonCerrar = tk.Button(master=raiz, text="Cerrar", command=cerrar)
#Se empaqueta en su contenedor que es raiz
botonCerrar.pack(side=tk.BOTTOM)

#La función para el boton con un «->» de la barraHerramientas
def botonSiguiente():
    #Aquí entra la variable «contadorPantallas»
    ##Pero, como es una variable global hay que especificarlo
    global contadorPantallas

    #Como se le dio clic anterior la «pantalla» ya no es la n, sino la n+1 y así con cada clic    
    ##Antes, hay que la ubicación no sea la ultima pantalla o de lo contrario no se puede avanzar
    if contadorPantallas < 2:
        contadorPantallas+=1

        #Dependiendo del contadorPantallas se nos muestre un texto distinto en el label que se creó 
        if contadorPantallas == 0:
            explicacion.config(text=textoCero)
            primerGrafica()
            
        elif contadorPantallas == 1:
            explicacion.config(text=textoUno)
            segundaGrafica()
            
        elif contadorPantallas == 2:
            explicacion.config(text=textoDos)
            tercerGrafica()    

#La función para el boton con un «<-» de la barraHerramientas
def botonAnterior():
    #De nuevo entra la variable «contadorPantallas»
    ##Pero, como es una variable global hay que especificarlo
    global contadorPantallas

    #Como se le dio clic anterior la «pantalla» ya no es la n, sino la n-1 y así con cada clic    
    ##Antes, hay que la ubicación no sea la pantalla cero o de lo contrario no se puede regresar
    if contadorPantallas > 0:
        contadorPantallas-=1

        #Dependiendo del contadorPantallas se nos muestre un texto distinto en el label que se creó 
        if contadorPantallas == 0:
            explicacion.config(text=textoCero)
            primerGrafica()
        elif contadorPantallas == 1:
            explicacion.config(text=textoUno)
            segundaGrafica()
            
        elif contadorPantallas == 2:
            explicacion.config(text=textoDos)
            tercerGrafica()

#Se modifica la función del buton1 de la barraHerramientas que es justo el botón con una ->
##De aquí en adelante se le llamara al ese botón siguiente
barraHerramientas.children['!button2'].config(command=botonAnterior)


#Se modifica la función del buton1 de la barraHerramientas que es justo el botón con una ->
##De aquí en adelante se le llamara al ese botón siguiente
barraHerramientas.children['!button3'].config(command=botonSiguiente)

tk.mainloop()
```
