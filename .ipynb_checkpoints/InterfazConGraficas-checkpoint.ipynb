```python
import tkinter as tk

from matplotlib.backends.backend_tkagg import (
    FigureCanvasTkAgg, NavigationToolbar2Tk)
#Implementar the las Matplotlib key bindings por defecto
from matplotlib.backend_bases import key_press_handler
from matplotlib.figure import Figure

import numpy as np

#El papel de ésta variable es «ubicar» en que paso del algoritmo se encuentra el usuario
##Más adelante en las funcines botonSiguiente y botonAnterior
contadorPantallas = 0

#Se crea la ventana principal que de aquí en adelante se le llamará raiz
raiz = tk.Tk()

#Se le agrega un título a la raíz
raiz.wm_title("Máquina de soporte vectorial")

#Se crea una figura con acho quince y altura cuatro
figura = Figure(figsize=(8, 4), dpi=100)

#Se parte en intervalo en partes muy pequeñas, esta lista de aquí en adelante se llamará equis
##Esto se usa para que más adelante la gráfica se vea continua
equis = np.arange(0, 3, .01)

#Se le agreda una subgráfica a la figura
##En el eje de las abscisas tendra a equis
##En el eje de las ordenadas tendra las imagen de la función aplicada a equis
figura.add_subplot(111).plot(equis, 2 * np.sin(2 * np.pi * equis))

#Se usa FigureCanvasTkAgg que importamos de matplotlib.backend_bases
##El canvas se agrega a la figura y esta se coloca en raiz
##canvas es una tk.DrawingArea, es decir, es un lugar para graficar
canvas = FigureCanvasTkAgg(figura, master=raiz)  #
#Se «imprime» el canvas
canvas.draw()
#Se agrega como un widget
canvas.get_tk_widget().pack(side=tk.TOP, fill=tk.BOTH, expand=1)

#Se le egrega una barra de herramientas ligada el canvas que estará en la raiz
##De aquí en adelante se le llamara barraHerramientas
##Para esto se usa NavigationToolbar2Tk que importamos de matplotlib.backends.backend_tkagg
barraHerramientas = NavigationToolbar2Tk(canvas, raiz)
barraHerramientas.update()
#Se agrega como un widget la barra de herramientas
canvas.get_tk_widget().pack(side=tk.TOP, fill=tk.BOTH, expand=1)

#Se va a usar una etiqueta en la ventana y con un texto que irá explicando el algoritmo
##El texto sera de color rojo, tipo de letra «Comic Sans MS» y tamaño 18
explicacion = tk.Label(raiz, text="algo", fg="white", font=("Comic Sans MS", 18))
##También lo ubicamos
explicacion.pack()
        
    
#Sinceramente, no se para que sirve está función, si alguien puede investigar y usarla me dice
def on_key_press(evento):
    print("you pressed {}".format(evento.key))
    key_press_handler(evento, canvas, barraHerramientas)

canvas.mpl_connect("key_press_event", on_key_press)


#Se define la función ligada con el botón para cerrar la ventana
def cerrar():
    raiz.quit()     # Para el mainloop
    raiz.destroy()  # Esto es necesario para prevenir el Fatal Python Error: PyEval_RestoreThread: NULL tstate

def my_function():
    print("Pressed Home")

#barraHerramientas.children['!button1'].config(command=my_function)
    
#Se crea el botón para cerrar la ventana
#Se ubica en al raíz, se el pone como texto «Cerrar» y usa la función cerrar
botonCerrar = tk.Button(master=raiz, text="Cerrar", command=cerrar)
#Se empaqueta en su contenedor que es raiz
botonCerrar.pack(side=tk.BOTTOM)

#La función para el boton con un «->» de la barraHerramientas
def botonSiguiente():
    #Aquí entra la variable «contadorPantallas»
    ##Pero, como es una variable global hay que especificarlo
    global contadorPantallas

    #Como se le dio clic anterior la «pantalla» ya no es la n, sino la n+1 y así con cada clic    
    ##Antes, hay que la ubicación no sea la ultima pantalla o de lo contrario no se puede avanzar
    if contadorPantallas < 2:
        contadorPantallas+=1

        #Dependiendo del contadorPantallas se nos muestre un texto distinto en el label que se creó 
        if contadorPantallas == 0:
            explicacion.config(text="texto cero")
       
        elif contadorPantallas == 1:
            explicacion.config(text="texto uno")
        
        elif contadorPantallas == 2:
            explicacion.config(text="texto dos")
                

#La función para el boton con un «<-» de la barraHerramientas
def botonAnterior():
    #De nuevo entra la variable «contadorPantallas»
    ##Pero, como es una variable global hay que especificarlo
    global contadorPantallas

    #Como se le dio clic anterior la «pantalla» ya no es la n, sino la n-1 y así con cada clic    
    ##Antes, hay que la ubicación no sea la pantalla cero o de lo contrario no se puede regresar
    if contadorPantallas > 0:
        contadorPantallas-=1

        #Dependiendo del contadorPantallas se nos muestre un texto distinto en el label que se creó 
        if contadorPantallas == 0:
            explicacion.config(text="texto cero")
       
        elif contadorPantallas == 1:
            explicacion.config(text="texto uno")
        
        elif contadorPantallas == 2:
            explicacion.config(text="texto dos")
        

#Se modifica la función del buton1 de la barraHerramientas que es justo el botón con una ->
##De aquí en adelante se le llamara al ese botón siguiente
barraHerramientas.children['!button2'].config(command=botonAnterior)


#Se modifica la función del buton1 de la barraHerramientas que es justo el botón con una ->
##De aquí en adelante se le llamara al ese botón siguiente
barraHerramientas.children['!button3'].config(command=botonSiguiente)

tk.mainloop()
```

```python
#La librería que se usará para realizar la interfaz se puede usar WxPython, PyQT, PyGTK y Tkinter
#Aquí se va a usar el módulo tkinter y como alias se le pondrá «tk»

import tkinter as tk

#Primeo se tiene que construir la ventana (o contenedor) principal, lo llamaremos raiz y dentro de ella podemos poner otros widgets

raiz = tk.Tk()

#El papel de ésta variable es «ubicar» en que paso del algoritmo se encuentra el usuario

contadorPantallas = 0

#Le vamos a dar título a la ventana

raiz.title("Clasificación lineal con MSV")

#Damos permiso para redimensionar la ventana y que el usuario no se vea incómodo
#Como parámetros primero esta el ancho y luego el alto

raiz.resizable(True,True)

#Damos un color de fondo

raiz.config(bg="red")

#Se crea ahora otro contenedor (un frame) específicando que va a estar en la ventana principal

primerFrame=tk.Frame(raiz)

#Luego, el frame se mete en la ventana principal, es decir, se empaqueta 

primerFrame.pack()

#Ahora, ponemos un color de fondo para el primerFrame

primerFrame.config(bg="blue")

#Para dale una dimensión inicial al primerFrame
#No se hace lo mismo con la raiz porque esta se ajusta a el tamaño de sus componentes

primerFrame.config(width="650", height="350")

#Se agrega un borde de tamaño 35 y tipo groove al primerFrame

primerFrame.config(bd=35)

primerFrame.config(relief="groove")

#También hacemos que sobre el contenerdor el cursos cambie

primerFrame.config(cursor="hand2")

#También le ponemos borde a la raiz

raiz.config(bd=35)

raiz.config(relief="groove")

#Se va a usar una etiqueta en primerFrame y con un texto introductivo a SVM
#El texto sera de color rojo, tipo de letra «Comic Sans MS» y tamaño 18

primerEtiqueta = tk.Label(primerFrame, text="algo", fg="white", font=("Comic Sans MS", 18))

#Posicionamos la etiqueta del lado izquierdo 

primerEtiqueta.grid(row=0, column=0)

#Creamos una segunda etiqueta para la imagen 
#Además, creamos una variable primerImagen de antemano para después poder usarla en la inicialización de la etiqueta

#primerImagen = tk.PhotoImage(file="icono.png")

#egundaEtiqueta = tk.Label(primerFrame, image=primerImagen)

#Empaquetamos la etiqueta 

#egundaEtiqueta.pack()

#Ahora vamos a realizar un botón que nos ayude a visualizar cual sería el «siguiente» paso de los algoritmos

primerBoton = tk.Button(primerFrame, text="Siguiente", command=lambda:funcionPrimerBoton())
#Posicionamos al botón 

primerBoton.grid(row=1, column=0)

#Ahora definimos la función del primer botón 

def funcionPrimerBoton():
    #Aquí es donde entra la variable «contadorPantallas»
    #como se le dio clic al botón de «siguiente» la «pantalla» ya no es la cero, sino la uno y así con cada clic
    global contadorPantallas
    
    contadorPantallas+=1
    
    #Todo va a entrar a  para que dependiendo del contadorPantallas se nos muestre un texto y gráfica distintos 
    if contadorPantallas == 0:
        primerEtiqueta.config(text="texto cero")
    elif contadorPantallas == 1:
        primerEtiqueta.config(text="texto uno")
    elif contadorPantallas == 2:
        primerEtiqueta.config(text="texto dos")
    
#Análogamente, vamos a realizar un botón que nos ayude a visualizar cual sería el paso «anterior» de los algoritmos
#Nota: se pone como una función lambda con el objetivo de que no se ejecute utomáticamente al iniciar el programa

botonAnterior = tk.Button(primerFrame, text="Anterior", command=lambda:funcionBotonAnterior())

#Posicionamos al botón «anterior»

botonAnterior.grid(row=1, column=1)

#Ahora definimos la función del primer botón 

def funcionBotonAnterior():
    #Aquí es donde entra la variable «contadorPantallas», pero se debe especificar que es una variable global
    global contadorPantallas
    
    
    #Como se le dio clic al botón de «anterior» la «pantalla» ya no es la cero, sino la uno y así con cada clic
    contadorPantallas-=1
    
    #Todo va a entrar a  para que dependiendo del contadorPantallas se nos muestre un texto y gráfica distintos 
    if contadorPantallas == 0:
        primerEtiqueta.config(text="texto cero")
    elif contadorPantallas == 1:
        primerEtiqueta.config(text="texto uno")
    elif contadorPantallas == 2:
        primerEtiqueta.config(text="texto dos")
    

    
#Por último, usamos el método mainLoop(), es para que la ventana funcione en un bucle infinito

raiz.mainloop()
```
