```python
#La librería que se usará para realizar la interfaz se puede usar WxPython, PyQT, PyGTK y Tkinter
#Aquí se va a usar el módulo tkinter y como alias se le pondrá «tk»

import tkinter as tk

#Primeo se tiene que construir la ventana (o contenedor) principal, lo llamaremos raiz y dentro de ella podemos poner otros widgets

raiz = tk.Tk()

#El papel de ésta variable es «ubicar» en que paso del algoritmo se encuentra el usuario

contadorPantallas = 0

#Le vamos a dar título a la ventana

raiz.title("Clasificación lineal con MSV")

#Damos permiso para redimensionar la ventana y que el usuario no se vea incómodo
#Como parámetros primero esta el ancho y luego el alto

raiz.resizable(True,True)

#Damos un color de fondo

raiz.config(bg="red")

#Se crea ahora otro contenedor (un frame) específicando que va a estar en la ventana principal

primerFrame=tk.Frame(raiz)

#Luego, el frame se mete en la ventana principal, es decir, se empaqueta 

primerFrame.pack()

#Ahora, ponemos un color de fondo para el primerFrame

primerFrame.config(bg="blue")

#Para dale una dimensión inicial al primerFrame
#No se hace lo mismo con la raiz porque esta se ajusta a el tamaño de sus componentes

primerFrame.config(width="650", height="350")

#Se agrega un borde de tamaño 35 y tipo groove al primerFrame

primerFrame.config(bd=35)

primerFrame.config(relief="groove")

#También hacemos que sobre el contenerdor el cursos cambie

primerFrame.config(cursor="hand2")

#También le ponemos borde a la raiz

raiz.config(bd=35)

raiz.config(relief="groove")

#Se va a usar una etiqueta en primerFrame y con un texto introductivo a SVM
#El texto sera de color rojo, tipo de letra «Comic Sans MS» y tamaño 18

primerEtiqueta = tk.Label(primerFrame, text="algo", fg="white", font=("Comic Sans MS", 18))

#Posicionamos la etiqueta del lado izquierdo 

primerEtiqueta.grid(row=0, column=0)

#Creamos una segunda etiqueta para la imagen 
#Además, creamos una variable primerImagen de antemano para después poder usarla en la inicialización de la etiqueta

primerImagen = tk.PhotoImage(file="icono.png")

#egundaEtiqueta = tk.Label(primerFrame, image=primerImagen)

#Empaquetamos la etiqueta 

#egundaEtiqueta.pack()

#Ahora vamos a realizar un botón que nos ayude a visualizar cual sería el «siguiente» paso de los algoritmos

primerBoton = tk.Button(primerFrame, text="Siguiente", command=lambda:funcionPrimerBoton())
#Posicionamos al botón 

primerBoton.grid(row=1, column=0)

#Ahora definimos la función del primer botón 

def funcionPrimerBoton():
    #Aquí es donde entra la variable «contadorPantallas»
    #como se le dio clic al botón de «siguiente» la «pantalla» ya no es la cero, sino la uno y así con cada clic
    global contadorPantallas
    
    contadorPantallas+=1
    
    #Todo va a entrar a  para que dependiendo del contadorPantallas se nos muestre un texto y gráfica distintos 
    if contadorPantallas == 0:
        primerEtiqueta.config(text="texto cero")
    elif contadorPantallas == 1:
        primerEtiqueta.config(text="texto uno")
    elif contadorPantallas == 2:
        primerEtiqueta.config(text="texto dos")
    
#Análogamente, vamos a realizar un botón que nos ayude a visualizar cual sería el paso «anterior» de los algoritmos
#Nota: se pone como una función lambda con el objetivo de que no se ejecute utomáticamente al iniciar el programa

botonAnterior = tk.Button(primerFrame, text="Anterior", command=lambda:funcionBotonAnterior())

#Posicionamos al botón «anterior»

botonAnterior.grid(row=1, column=1)

#Ahora definimos la función del primer botón 

def funcionBotonAnterior():
    #Aquí es donde entra la variable «contadorPantallas», pero se debe especificar que es una variable global
    global contadorPantallas
    
    
    #Como se le dio clic al botón de «anterior» la «pantalla» ya no es la cero, sino la uno y así con cada clic
    contadorPantallas-=1
    
    #Todo va a entrar a  para que dependiendo del contadorPantallas se nos muestre un texto y gráfica distintos 
    if contadorPantallas == 0:
        primerEtiqueta.config(text="texto cero")
    elif contadorPantallas == 1:
        primerEtiqueta.config(text="texto uno")
    elif contadorPantallas == 2:
        primerEtiqueta.config(text="texto dos")
    

    
#Por último, usamos el método mainLoop(), es para que la ventana funcione en un bucle infinito

raiz.mainloop()
```
